```python
# === arrival_matrix_batch_save.py ===
# (Or paste this into a fresh notebook cell.)

import os
import glob
import cv2
import numpy as np

# ─────────────────────────────────────────────────────────────────────────────
# 1) CONFIGURATION
# ─────────────────────────────────────────────────────────────────────────────

# List here all of the image folders you want to process:
VERSION_FOLDERS = [

    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Wool = Leaves\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Wool = Leaves\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Wool = Leaves\v3",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Wool = Leaves\v4",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Wool = Leaves\v5",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Wool = Leaves\v6",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Wool = Leaves\v7",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Wool = Leaves\v8",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Wool = Leaves\v9",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Wool = Leaves\v10",

    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V4 Remove Adj MaxAge Fire\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V4 Remove Adj MaxAge Fire\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V4 Remove Adj MaxAge Fire\v3",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V4 Remove Adj MaxAge Fire\v4",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V4 Remove Adj MaxAge Fire\v5",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V4 Remove Adj MaxAge Fire\v6",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V4 Remove Adj MaxAge Fire\v7",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V4 Remove Adj MaxAge Fire\v8",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V4 Remove Adj MaxAge Fire\v9",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V4 Remove Adj MaxAge Fire\v10",



    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V3 No Adjacent Block Removal\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V3 No Adjacent Block Removal\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Green Leaves - V3 No Adjacent Block Removal\v3",

    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - V2 No Random Fire Leave\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - V2 No Random Fire Leave\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - V2 No Random Fire Leave\v3",
    
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - V3 No Adjacent Block Removal\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - V3 No Adjacent Block Removal\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - V3 No Adjacent Block Removal\v3",

    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - V3.5 Keep the random removal of fire chance\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - V3.5 Keep the random removal of fire chance\v2",



    
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\Sample 100% Spread Chance",

    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\HALF, Low AE+High BT v High AE+Low BT\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\HALF, Low AE+High BT v High AE+Low BT\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\HALF, Low AE+High BT v High AE+Low BT\v3",

    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\HALF BT = 50, Low AE v High AE\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\HALF BT = 50, Low AE v High AE\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\HALF BT = 50, Low AE v High AE\v3",

    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\HALF AE = 50, Low BT v High BT\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\HALF AE = 50, Low BT v High BT\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\HALF AE = 50, Low BT v High BT\v3",




    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves\v3",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves\v4",
    r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves\v5",

    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M2 New Fire Age\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M2 New Fire Age\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M2 New Fire Age\v3",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M2 New Fire Age\v4",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M2 New Fire Age\v5",

    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M3 Linearise Fire Age\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M3 Linearise Fire Age\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M3 Linearise Fire Age\v3",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M3 Linearise Fire Age\v4",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M3 Linearise Fire Age\v5",

    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M4 Remove Adj MaxAge Fire\v1",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M4 Remove Adj MaxAge Fire\v2",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M4 Remove Adj MaxAge Fire\v3",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M4 Remove Adj MaxAge Fire\v4",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M4 Remove Adj MaxAge Fire\v5",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M4 Remove Adj MaxAge Fire\v6",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M4 Remove Adj MaxAge Fire\v7",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M4 Remove Adj MaxAge Fire\v8",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M4 Remove Adj MaxAge Fire\v9",
    # r"C:\Users\samih\Documents\ShareX\Screenshots\For VS Vision\34 x 34\QUAD, 1 Log, 2 Bush, 3 Hay, 4 Leaves - M4 Remove Adj MaxAge Fire\v10",
]

GRID_SIZE           = 34
IMAGE_STEP          = 1   # process every image; set to e.g. 5 to subsample
TIME_STEP           = 1   # seconds between frames
OUTPUT_DIR          = "Arrival Matrices"
os.makedirs(OUTPUT_DIR, exist_ok=True)


# ─────────────────────────────────────────────────────────────────────────────
# 2) CORE FUNCTIONS (only what we need)
# ─────────────────────────────────────────────────────────────────────────────

def classify_cell_color(avg_bgr):
    """Exactly your HSV‐threshold classifier."""
    b, g, r = avg_bgr
    color_bgr = np.uint8([[avg_bgr]])
    h, s, v = cv2.cvtColor(color_bgr, cv2.COLOR_BGR2HSV)[0,0]
    if (30 <= h <= 90)  and (s > 50) and (v > 50):   return 'green'
    if (130<= h <=160) and (s > 70) and (v > 50):   return 'purple'
    if (160<= h <=180) and (s > 50) and (v > 100):  return 'pink'
    if (80 <= h <=100) and (s > 50) and (v > 80):   return 'lightblue'
    if (0  <= h <=40)  and (s > 60) and (v > 60):   return 'orange'
    return 'white'

def process_image_into_grid(img, grid_size=GRID_SIZE):
    """
    Split img into grid_size^2 cells, classify each, return label‐grid.
    (We only need the labels to detect 'orange'.)
    """
    h, w, _ = img.shape
    cell_h, cell_w = h//grid_size, w//grid_size

    labels = []
    for row in range(grid_size):
        row_labels = []
        for col in range(grid_size):
            # border forced black (won’t turn orange)
            if row in (0,grid_size-1) or col in (0,grid_size-1):
                row_labels.append('black')
            else:
                y0,y1 = row*cell_h, (row+1)*cell_h
                x0,x1 = col*cell_w, (col+1)*cell_w
                cell = img[y0:y1, x0:x1]
                avg = cell.reshape(-1,3).mean(axis=0)  # BGR
                row_labels.append(classify_cell_color(avg))
        labels.append(row_labels)
    return np.array(labels)

def load_label_matrices(folder, grid_size=GRID_SIZE, image_step=IMAGE_STEP):
    """Read all PNG/JPGs, process into label grids, return list."""
    paths = sorted(glob.glob(os.path.join(folder, "*.png")), key=os.path.getmtime)
    if not paths:
        paths = sorted(glob.glob(os.path.join(folder, "*.jpg")), key=os.path.getmtime)
    paths = paths[::image_step]

    mats = []
    for p in paths:
        img = cv2.imread(p)
        if img is None:
            print("⚠️ failed to load:", p)
            continue
        mats.append(process_image_into_grid(img, grid_size))
    return mats

def compute_arrival_time_matrix(label_mats, time_step=TIME_STEP):
    """
    From a sequence of label‐grids, build arrival_matrix[i,j] = first t when 'orange'.
    """
    N = label_mats[0].shape[0]
    arr = np.full((N,N), np.inf, dtype=float)

    for t, mat in enumerate(label_mats):
        mask = (mat == 'orange')
        arr[mask & np.isinf(arr)] = t * time_step

    arr[np.isinf(arr)] = np.nan
    return arr
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# 3) BATCH LOOP: compute & SAVE
# ─────────────────────────────────────────────────────────────────────────────

for folder in VERSION_FOLDERS:
    print("Processing:", folder)
    labels = load_label_matrices(folder, grid_size=GRID_SIZE, image_step=IMAGE_STEP)
    arrival = compute_arrival_time_matrix(labels, time_step=TIME_STEP)

    # build a clean filename from the part after "For VS Vision\"
    key = folder.split("For VS Vision\\")[-1].replace("\\","--")
    fname = f"{key} arrival.npy"
    outpath = os.path.join(OUTPUT_DIR, fname)

    np.save(outpath, arrival)
    print(" → saved:", outpath)

print("All done.  Arrival matrices are in:", OUTPUT_DIR)
```
